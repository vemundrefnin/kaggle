```python
import pandas as pd
```

```python
train_data = pd.read_csv('data/train.csv')
test_data = pd.read_csv('data/test.csv')
stores = pd.read_csv('data/stores.csv')
oil = pd.read_csv('data/oil.csv')
holiday = pd.read_csv('data/holidays_events.csv')
transactions = pd.read_csv('data/transactions.csv')
```

```python

train_data_combined = train_data.merge(stores, on='store_nbr')
test_data_combined = test_data.merge(stores, on='store_nbr')

train_data.shape
```

```python
train_data_combined['date'] = pd.to_datetime(train_data_combined['date'])
train_data_combined['year'] = train_data_combined['date'].dt.year
train_data_combined['month'] = train_data_combined['date'].dt.month
train_data_combined['day'] = train_data_combined['date'].dt.day
```

```python
test_data_combined['date'] = pd.to_datetime(test_data_combined['date'])
test_data_combined['year'] = test_data_combined['date'].dt.year
test_data_combined['month'] = test_data_combined['date'].dt.month
test_data_combined['day'] = test_data_combined['date'].dt.day
```

```python
train_data_combined.to_csv("data/train_data_combined.csv")
test_data_combined.to_csv("data/test_data_combined.csv")
```
