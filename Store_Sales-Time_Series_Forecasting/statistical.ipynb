```python
from sklearn import linear_model
#from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('data/train_data_combined.csv')
test = pd.read_csv("data/test_data_combined.csv")
```

```python
features = ["year", "month", "day", "cluster"]
train["cluster"] = train.cluster.astype("category")
test["cluster"] = test.cluster.astype("category")
```

```python
x = train[features]
y = train['sales']
x= pd.get_dummies(data=x, drop_first=True)
x
```

```python
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(x, y)
```

```python
x_test = test[features]
x_test= pd.get_dummies(data=x, drop_first=True)
x_test
```

```python
estiamtes = regr.predict(x_test)
```

```python
plt.plot(train.sales)
```

```python
plt.plot(estiamtes)
```

```python
test["sales"] = pd.Series(estiamtes)
```
